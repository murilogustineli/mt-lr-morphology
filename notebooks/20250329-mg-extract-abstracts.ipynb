```python
import pandas as pd
from gpt.config import get_data_path


# data path
base_path = get_data_path()
file_path = "data/dataset-updated.xlsx"
data_path = f"{base_path}/{file_path}"
print(data_path)

# read data
df = pd.read_excel(data_path, engine="openpyxl")
print(f"Number of rows: {len(df)}")
df.head(3)
```

```python
# filter for True abstracts in status column
# Convert string "True"/"False" to actual booleans, if needed
df["status"] = df["status"].astype(str)
df_abs = df[df["status"] == "True"]
df_abs.head(3)
```

```python
df_abs["status"].value_counts()
```

```python
# write to excel
output_path = f"{base_path}/data/abstracts.xlsx"
df_abs.to_excel(output_path, index=False)
print(f"Filtered abstracts saved to: {output_path}")
```

```python
# read the saved file to verify
df_check = pd.read_excel(output_path, engine="openpyxl")
print(f"Number of rows in the saved file: {len(df_check)}")
df_check.head(3)
```
